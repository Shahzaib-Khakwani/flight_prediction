```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_train = pd.read_csv("content/sastatickets/sastaticket_train.csv")
df_test = pd.read_csv("content/sastatickets/sastaticket_test.csv")
df_train.head()
```

```python
df_train.isnull().sum()
```

```python
df_train.info()
```

```python
df_train.drop(['Unnamed: 0.2', 'Unnamed: 0', 'f10', 'f2', 'f3'],axis= 1, inplace= True)
```

```python
df_train.head()
```

```python
columns = df_train.columns
for i in columns[3:6]:
    print(i, df_train[i].unique())
```

```python
df_train['f1'] = pd.to_datetime(df_train['f1'])
df_train['f4'] = pd.to_datetime(df_train['f4'])
df_train['f5'] = pd.to_datetime(df_train['f5'])
```

```python
df_train.info()
```

```python
df_train.insert(0, "to_departure(s)", (df_train.f4 - df_train.f1).astype('timedelta64[s]'))
df_train.insert(1, "travel_time(s)", (df_train.f5 - df_train.f4).astype('timedelta64[s]'))
```

```python
df_train.head()
```

```python
df_train.drop(['f1', 'f4', 'f5'], axis=1, inplace = True)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in columns[3:6]:
    df_train[i] = le.fit_transform(df_train[i])
    
df_train.head()
```

```python
#standardScaler
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
X = df_train.drop(["target", "Unnamed: 0.1"], axis = 1)
y = df_train['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 6)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
knr = KNeighborsRegressor()
```

```python
for i in [lr, dt, svr, knr]:
    i.fit(X_train, y_train)
    train_pred = i.predict(X_train)
    test_pred = i.predict(X_test)
    print(i)
    train_accuracy = r2_score(y_train, train_pred)
    test_accuracy = r2_score(y_test, test_pred)
    print("train accuracy", train_accuracy, mean_absolute_error(y_train, train_pred))
    print("test accuracy", test_accuracy, mean_absolute_error(y_test, test_pred))
    print('---------------------')
    
    
    
```
